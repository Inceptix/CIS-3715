```python
NY_state_all = df[(df['sub_region_1']=='New York')] #163076
NY_state = NY_state_all[:57] #index_last - index_first + 1
NY_state['counts'] = np.arange(len(NY_state))
```

```python
plt.plot(NY_state['counts'], NY_state['grocery_and_pharmacy_percent_change_from_baseline'], marker='', color='red', linewidth=3, alpha=0.7)
plt.plot(NY_state['counts'], NY_state['retail_and_recreation_percent_change_from_baseline'], marker='', color='green', linewidth=3, alpha=0.7)
```

```python
NY_county = NY_state_all[1710:1767] #index_last - index_first + 1
NY_county['counts'] = np.arange(len(NY_county))
```

```python
plt.figure(figsize =(14, 8))
plt.title('New York Retail and Recreation Mobility Trend')
plt.ylabel('Percentage')
plt.xticks(rotation=90)
plt.grid()

plt.plot(NY_county['date'], NY_county['grocery_and_pharmacy_percent_change_from_baseline'], marker='', color='blue', linewidth=3, alpha=0.7)
plt.plot(NY_county['date'], NY_county['retail_and_recreation_percent_change_from_baseline'], marker='', color='red', linewidth=3, alpha=0.7)
plt.plot(NY_state['date'], NY_state['grocery_and_pharmacy_percent_change_from_baseline'], marker='', color='blue', linewidth=3, alpha=0.7)
plt.plot(NY_state['date'], NY_state['retail_and_recreation_percent_change_from_baseline'], marker='', color='red', linewidth=3, alpha=0.7)

plt.text(57, NY_state.grocery_and_pharmacy_percent_change_from_baseline.tail(1),'New York State',horizontalalignment='left',size='large' ,color='blue')
plt.text(57, NY_state.retail_and_recreation_percent_change_from_baseline.tail(1),'New York State',horizontalalignment='left',size='large' ,color='red')
plt.text(57, NY_county.grocery_and_pharmacy_percent_change_from_baseline.tail(1),'New York County',horizontalalignment='left',size='large' ,color='blue')
plt.text(57, NY_county.retail_and_recreation_percent_change_from_baseline.tail(1),'New York County',horizontalalignment='left',size='large' ,color='red')

plt.text(45, 22,'---- Grocery and Pharmacy',horizontalalignment='left',size='large' ,color='blue')
plt.text(45, 18,'---- Retail and Recreation',horizontalalignment='left',size='large' ,color='red')
```

```python
ITA = df[(df['country_region']=='Italy')] #39790

ITA_avg = ITA[( ITA['sub_region_1'].isnull())]  #39790
ITA_avg['counts'] = np.arange(len(ITA_avg))

ITA_piedmont = ITA[(ITA['sub_region_1']=='Piedmont')] #40672
ITA_piedmont['counts'] = np.arange(len(ITA_piedmont))

ITA_lomb = ITA[(ITA['sub_region_1']=='Lombardy')]
ITA_lazio = ITA[(ITA['sub_region_1']=='Lazio')]
```

```python
plt.figure(figsize =(14, 8))
plt.title('Italy Retail and Recreation Mobility Trend')
plt.ylabel('Percentage')
plt.xticks(rotation=90)
plt.grid()

plt.plot(ITA_lomb['date'], ITA_lomb['retail_and_recreation_percent_change_from_baseline'], marker='', color='red', linewidth=3, alpha=0.7)
plt.plot(ITA_lazio['date'], ITA_lazio['retail_and_recreation_percent_change_from_baseline'], marker='', color='blue', linewidth=3, alpha=0.7)

plt.text(63, ITA_lomb.retail_and_recreation_percent_change_from_baseline.tail(1),'Italy - Lmbardy Region',horizontalalignment='left',size='large' ,color='red')
plt.text(63, ITA_lazio.retail_and_recreation_percent_change_from_baseline.tail(1),'Italy - Lazio Region',horizontalalignment='left',size='large' ,color='blue')
```

```python
plt.figure(figsize =(14, 8))
plt.title('Italy Grocery and Pharmacy Mobility Trend')
plt.ylabel('Percentage')
plt.xticks(rotation=90)
plt.grid()

plt.plot(ITA_lomb['date'], ITA_lomb['grocery_and_pharmacy_percent_change_from_baseline'], marker='', color='red', linewidth=3, alpha=0.7)
plt.plot(ITA_lazio['date'], ITA_lazio['grocery_and_pharmacy_percent_change_from_baseline'], marker='', color='blue', linewidth=3, alpha=0.7)

plt.text(63, ITA_lomb.grocery_and_pharmacy_percent_change_from_baseline.tail(1),'Italy - Lmbardy Region',horizontalalignment='left',size='large' ,color='red')
plt.text(63, ITA_lazio.grocery_and_pharmacy_percent_change_from_baseline.tail(1),'Italy - Lazio Region',horizontalalignment='left',size='large' ,color='blue')
```

```python
plt.figure(figsize =(14, 8))
plt.title('Italy Mobility Trend')
plt.ylabel('Percentage')
plt.xticks(rotation=90)
plt.grid()

plt.plot(ITA_avg['date'], ITA_avg['grocery_and_pharmacy_percent_change_from_baseline'], marker='', color='red', linewidth=3, alpha=0.7)
plt.plot(ITA_avg['counts'], ITA_avg['retail_and_recreation_percent_change_from_baseline'], marker='', color='purple', linewidth=3, alpha=0.7)
plt.plot(ITA_avg['counts'], ITA_avg['transit_stations_percent_change_from_baseline'], marker='', color='blue', linewidth=3, alpha=0.7)
plt.plot(ITA_avg['counts'], ITA_avg['workplaces_percent_change_from_baseline'], marker='', color='green', linewidth=3, alpha=0.7)

plt.text(63, ITA_avg.grocery_and_pharmacy_percent_change_from_baseline.tail(1),'Grocery and Pharmacy',horizontalalignment='left',size='large' ,color='red')
plt.text(63, ITA_avg.retail_and_recreation_percent_change_from_baseline.tail(1),'Retail and Recreation',horizontalalignment='left',size='large' ,color='purple')
plt.text(63, ITA_avg.transit_stations_percent_change_from_baseline.tail(1),'Public Transit',horizontalalignment='left',size='large' ,color='blue')
plt.text(63, ITA_avg.workplaces_percent_change_from_baseline.tail(1),'Workplace',horizontalalignment='left',size='large' ,color='green')
```
