Using kNN and decision trees on the problem of churn prediction.

This is expensive for the company, as a lost customer means lost monthly revenue.

Lets assume we have to use the churn classifier to contact your existing customers and try to prevent churn. The cost of each contact is `$10`. Let us assume each contacted customer will decide to stay. The profit in keeping a customer is `$30`.

The kNN classifier has a 88.40% accuracy. The classifier has 96 false negative cases which means `96 x -30 = 2880 dollars` are lost because the model incorrectly predicted the customer will not churn. It also has 20 false positives which means the company would unnecessary spend `20 x -10 = 200 dollars` on customers who will not churn. The model correctly predicts that 843 customers will not leave so we do not spend money on themfor a profit of `843 x 30 = 25,290 dollars` and predict correctly that 41 will churn so on those we have `41 x 20 = 820 dollars.` The total profit with the kNN classifier is `23,030 dollars`. 

The decision tree classifier has 94.2% accuracy. Furthermore looking at the confusion matrix we have way less flase positives and false negatives. There are 38 cases of false positives where we end up spending unecessary money with contacting the costumer. `38 x -10 = -380 dollars.` There are 27 false negatives where we do not spend contact money but lose the costumer. `27 x -30 = -810 dollars`. We predict correctly that 832 costumers will stay `831 x 30 = 24,930 dollars` and predict correctly that 110 customers will churn so we spend 10 dollars per contact to gain 30 for a net gain of 20 dollars per customer. `110 x 20 = 2,200 dollars.` The total profit with this decision tree classifier is `25,940 dollars`.
