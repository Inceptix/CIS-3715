```python
import pandas as pd
import numpy as np
import re
pd.set_option('display.expand_frame_repr', False) #prevents multiple lines of pandas data frames
```

```python
df1 = pd.read_csv('defect_report.csv')  # df is DataFrame object
df2 = pd.read_csv('defect_words.csv')
```

```python
word_list = df2['words'].tolist()
word_string = '(david)|(henry)|(stanley)|(glass)|(\\band\\b)|(before)'
counter = 0
df1['description'] = df1['description'].str.lower()
```

```python
word_join = ('|'.join('(\\b{}\\b)'.format(i) for i in word_list))
word_join
```

```python
defects = df1['description'].str.count(word_join, flags=re.I)
```

```python
defects
```

```python
df1['defect'] = defects
df1
```

```python
df1['defect'] = df1['description'].str.contains(word_string, case=False, regex=True)
df1
```
